```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
n = [1,4,16,64]
```

```python
x = np.arange(-5,5,0.1)
```

```python
def sigmoid(x,n=1):
    return np.exp(n*x)/(1+np.exp(n*x))
```

```python
df = pd.DataFrame({'GreyScale':x,'Acceptability n=1': sigmoid(x,n=1),'Acceptability n=4': sigmoid(x,n=4),\
                 'Acceptability n=16': sigmoid(x,n=16)}).set_index('GreyScale')
```

```python
lines = df.plot.line()
```
